```python
import numpy as np

valovna_dolžina=633#nm

print("________________________Postavitev__________________________________")
f1=float(input("Goriščna razdajlja prve leče v mm (f1):")) #goriščna razdajlja prve leče v mm
f2= float(input("Goriščna razdajlja druge leče v mm (f2):"))#goriščna razdalja druge leče v mm
f3= float(input("Goriščna razdajlja fokusne leče v mm (f3):"))#goriščna razdalja druge leče v mm
s1= float(input("Oddaljenost vhodnega žarka v mm (s1):"))#oddaljenost vhodnega žarka
print("______________________Vhodni_žarek__________________________________")
w01=float(input("POLMER vhodnega žarka v mm (w01):")) #POLMER vhodnega žarka
ZR1= float(input("ZR1 v mm:"))#mm
O1= float(input("O1 v mrad:"))#mrad

#___________________________________________________________________________
# f1=40 #goriščna razdajlja prve leče v mm
# f2= 120#goriščna razdalja druge leče v mm
# f3= 150#goriščna razdalja fokusne leče v mm
# w01= 0.335#POLMER vhodnega žarka
# s1= 40 #oddaljenost vhodnega žarka
# ZR1= 558#mm
# O1= 0.601#mrad
#___________________________________________________________________________
def s_izhod(f,s,ZR):
    return 1/(1/f-(1/(s+(ZR**2)/(s-f))))

def w_izhod(w_vhod,s,f,ZR):
    return w_vhod/(np.sqrt((1-s/f)**2+(ZR/f)**2))

def ZR_izhod(ZR,w_in,w_out):
    return ZR*(w_out/w_in)**2

print("____________________Žarek_med_lečama_________________________________")
w01_out= w_izhod(w01,s1,f1,ZR1)
w02=w01_out
print(f"w01\"=w02: {w01_out:.5f} mm")
if s1==f1:
    s2=f2
    print(f"s1\": {f1:.5f} mm")
    print(f"s2: {f2:.5f} mm")
else:
    s1_out= s_izhod(f1,s1,ZR1)
    print(f"s1\": {s1_out:.5f} mm")
    s2=f1+f2-s1_out
    print(f"s2: {s2:.5f} mm")
ZR1_out=ZR_izhod(ZR1,w01,w01_out)
ZR2=ZR1_out
print(f"ZR1\"=ZR2= {ZR1_out:.5f} mm")
print("______________________Izhodni_žarek__________________________________")
w02_out= w_izhod(w02,s2,f2,ZR2)
print(f"w02\": {w02_out:.5f} mm")
if s2==f2:
    s2_out=s2
    print(f"s2\": {f2:.5f} mm")
else:
    s2_out= s_izhod(f2,s2,ZR2)
    print(f"s2\": {s2_out:.5f} mm")
ZR2_out=ZR_izhod(ZR2,w02,w02_out)
print(f"ZR2\"= {ZR2_out:.5f} mm")
O2= valovna_dolžina/(np.pi*w02_out*1000)
print(f"O2\"= {O2:.5f} mrad")
print("______________________Fokusni_žarek__________________________________")
w03=w02_out
ZR3=ZR2_out
s3=f3+f2-s2_out
w03_out= w_izhod(w03,s3,f3,ZR3)
print(f"w03\": {w03_out:.5f} mm")
s3_out= s_izhod(f3,s3,ZR3)
print(f"s3\": {s3_out:.5f} mm")
ZR3_out=ZR_izhod(ZR3,w03,w03_out)
print(f"ZR3\"= {ZR3_out:.5f} mm")
O3= valovna_dolžina/(np.pi*w03_out*1000)
print(f"O2\"= {O2:.5f} mrad")
```
